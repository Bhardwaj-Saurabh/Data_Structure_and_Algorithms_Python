# **What is Big(0)?**

### In computer science and mathematics, Big O notation, also known as asymptotic notation, is used to describe the performance or complexity of an algorithm. It provides an upper bound on the growth rate of the algorithm as the input size increases. It helps us analyze how the algorithm's execution time or space requirements grow as the input size increases. It provides an upper bound estimate of the algorithm's scalability.

### The notation O(f(n)) represents the upper bound or worst-case scenario of an algorithm's time complexity, where "n" represents the input size and "f(n)" represents the growth rate function. In simpler terms, it tells you how the algorithm's performance scales with the input size.

### In Big O notation, we focus on the most significant factor that influences an algorithm's efficiency. It ignores constant factors and lower-order terms, focusing on the growth rate as the input size approaches infinity. It allows us to compare algorithms and understand their relative efficiency.

### For example, if an algorithm has a time complexity of O(n), where n is the input size, it means the algorithm's execution time grows linearly with the input size. If the input doubles, the execution time will also roughly double.

### To measure the worst-case time complexity of an algorithm, you typically analyze the algorithm and determine the maximum number of basic operations it performs as a function of the input size. This analysis helps identify the dominant operations or loops that contribute the most to the algorithm's time complexity.

# Here are some common worst-case time complexities and their notations:

### **O(1):** Constant time complexity. The algorithm's execution time remains the same regardless of the input size.

### **O(log n):** Logarithmic time complexity. The algorithm's execution time grows logarithmically with the input size.

### **O(n):** Linear time complexity. The algorithm's execution time grows linearly with the input size.

### **O(n log n):** Linearithmic time complexity. The algorithm's execution time grows in proportion to the product of the input size and the logarithm of the input size.

### **O(n^2):** Quadratic time complexity. The algorithm's execution time grows quadratically with the input size.

### **O(2^n):** Exponential time complexity. The algorithm's execution time grows exponentially with the input size.

### To determine the worst-case time complexity, you consider the most significant term in the number of operations performed by the algorithm. Additionally, you drop the lower-order terms and ignore constant factors.

### It's important to note that Big O notation provides an upper bound estimate and focuses on the worst-case scenario. It doesn't provide information about the best-case or average-case performance of an algorithm.


# **O(1)**

```python
'''
In this example, the function constant_example returns the first element of a list.
Regardless of the size of the list, it always takes a constant amount of time 
to access the first element. The execution time doesn't depend on the input size, 
so it has a constant time complexity.
'''

def constant_example(lst):
    return lst[0]
```

# **O(n)**

```python
'''
Here, the function linear_example iterates over each item in the list 
and performs a constant-time operation (printing) for each item. 
As the input list grows larger, the function takes proportionally 
more time to process each item. The execution time grows linearly 
with the input size, leading to a linear time complexity.
'''

def linear_example(lst):
    for item in lst:
        print(item)
```

# **O(log(n))**

```python
'''
In this example, the function logarithmic_example doubles the value of i 
in each iteration until it becomes greater than or equal to n. 
Since the value of i doubles at each step, the number of iterations 
required to reach n is roughly logarithmic to the input size n. 
The execution time grows logarithmically with the input size, 
resulting in a logarithmic time complexity.
'''

def logarithmic_example(n):
    i = 1
    while i < n:
        i *= 2
    return i
```

# **O(n^2)**

```python
'''
In this example, the function quadratic_example has nested loops that 
iterate over the list. For each item in the outer loop, 
the inner loop iterates over the entire list. As a result, 
the number of iterations becomes proportional to the square of the input size.
The execution time grows quadratically with the input size, 
resulting in a quadratic time complexity.
'''

def quadratic_example(lst):
    for item1 in lst:
        for item2 in lst:
            print(item1, item2)
```

# **O(2^n)**

```python
'''
n this example, the exponential_example function recursively calls itself twice 
with a reduced input size (n - 1) in each recursive call. 
As a result, the number of function calls doubles with each recursive level. 
The execution time grows exponentially with the input size n, 
resulting in an O(2^n) time complexity. 
This type of complexity is often associated with algorithms that solve problems 
by generating all possible subsets or combinations.
'''

def exponential_example(n):
    if n <= 1:
        return
    exponential_example(n - 1)
    exponential_example(n - 1)
```

# **O(n!)**

```python
'''
In this example, the factorial_example function recursively calculates 
the factorial of n. It multiplies n with the factorial of n-1 until 
it reaches the base case of n = 0. The number of recursive calls 
and the amount of computation grow rapidly as n increases. 
The execution time grows factorially with the input size n, 
resulting in an O(n!) time complexity. Algorithms with factorial complexity are 
generally highly inefficient and only feasible for very small inputs 
due to their exponential growth.
'''

def factorial_example(n):
    if n == 0:
        return 1
    return n * factorial_example(n - 1)

```

# **How to measure the performance of algorithms?**

### While calculating Big O notation for an algorithm, you can follow these simple rules:

### **Identify the Dominant Term:** Look for the most significant term that contributes the most to the algorithm's complexity. Focus on the operation that has the highest impact on the execution time or space requirements.

### **Eliminate Constants:** Ignore constant factors and coefficients. Big O notation focuses on the growth rate as the input size increases towards infinity. So, if an operation takes 2n or 3n iterations, we simply consider it as O(n).

### **Remove Lower-Order Terms:** Drop lower-order terms since they become insignificant as the input size grows. For example, if an algorithm has O(n^2 + n), we simplify it to O(n^2) because the n^2 term dominates the growth rate.

### **Combine Complexity of Nested Operations:** If an algorithm has nested loops or operations, you multiply the complexity of each operation. For example, if you have O(n) outer loop and O(m) inner loop, the combined complexity becomes O(n * m).

### **Choose the Worst Case:** Consider the scenario that results in the highest complexity. Big O notation focuses on the worst-case time complexity, representing the maximum number of operations an algorithm can take for any input.

### By following these rules, you can derive a simplified expression that represents the algorithm's time complexity using Big O notation. Keep in mind that Big O provides an upper bound estimate, allowing you to compare the scalability and efficiency of different algorithms as the input size grows.
